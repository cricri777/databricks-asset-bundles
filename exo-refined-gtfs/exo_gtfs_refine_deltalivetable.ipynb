```
import dlt
from pyspark.sql.functions import *

@dlt.view(comment="agency refined and cleanup")
def gtfs_refined_current_agency(env="dev"):
    agency_df = (
        spark.read.table(f"exo_{env}.gtfs_landing.agency")
        .where("landing_date = current_date()")

    )
    return agency_df
```
